The aim is to prove that in a cylindrical coordinate problem, the derivative of a property that is linearly corelated to the volume approaches zero when the same property is described as a function of the $r$ coordinate.

Mathematically:

If $f(V)= \alpha V$, then when the same property (described by $f(V)$) is described as a function of the $r$ coordinate, the derivative of this new function when $r$ approaches $0$ is $0$.

I.e. if $\frac{\partial f(V)}{\partial V} = A$ (with $A \in \mathbb{R}$), then $ \left|\frac{\partial f(r)}{\partial r}\right|_{r = 0} = 0$

A visual example is given below:

```python
import math
import numpy as np 
import matplotlib.pyplot as plt
```

```python
V_axis = np.linspace(0,100,10000)
y_axis = []
def f(V_axis,y_axis):
    for V in V_axis:
        y_axis.append(-V)
    return y_axis
y_axis = f(V_axis,y_axis)

plt.plot(V_axis,y_axis)
plt.show()

```

This is the graph of a function that is linear. 

$f(V) = \alpha x$ with alpha arbitrarily chosen to be $-1$ here.

Next, if the coordinate is changed from $V$ to $r$, the plot will ressemble:

```python
new_y_axis = y_axis
new_x_axis = []

def change_of_variable(V_axis,new_x_axis):
    for V in V_axis:
        new_x_axis.append(math.sqrt(V))
    return new_x_axis

new_x_axis = change_of_variable(V_axis,new_x_axis)

plt.plot(new_x_axis,new_y_axis)
plt.show()
```

Here, we can visually, see that the new function is no longer linear, but it is quadratic, i.e. $f(x) = A x^2 + B x + C $.

If we evaluate such a function, we see that the derivative at $x = 0$ is in fact $= 0$ when $B = 0$

We know that the temperature in the system is proportional to the volume.
(The temperature mentioned here is techincally the difference in temeprature between the inner timperature in the wire and the temperature at volume V or position r)

i.e 

$T(V) = \alpha V$

When writting this as a function of the radial distance in cylindrical units, this is given as:

$T(r) = \alpha \pi r^2$

This means that the derivative of the temeprature when $r$ tends to $0$ tends to $0$

i.e.

$\left|\frac{\partial T(r)}{\partial r}\right|_{r = 0} = \left|\frac{\partial \alpha \pi r^2}{\partial r}\right|_{r = 0} = \left| 2\alpha \pi r \right|_{r = 0} = 0$ 

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))


axes[0].plot(V_axis, y_axis, color='blue')
axes[0].set_title('T = f(V)')
axes[0].set_xlabel('V')
axes[0].set_ylabel('f(V)')  
axes[1].plot(new_x_axis, new_y_axis, color='red')
axes[1].set_title('T = f(r)')
axes[1].set_xlabel('r')
axes[1].set_ylabel('f(r)')
plt.show()
```
